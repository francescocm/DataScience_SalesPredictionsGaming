* En primer lugar, en los casos como Name y Genre con valores nulos muy bajo (2), eliminamos las filas con nulos en estas columnas.

* Para el a√±o de lanzamiento imputamos los valores nulos en la columna con la mediana y con la moda en la columna Publisher.

* Para el caso de las columnas Critic_Score, Critic_Count, User_Score, User_Count, Developer y Rating, dada la gran cantidad de valores nulos y que considero que no son esenciales para mi an√°lisis opto por eliminarlas.



### Tratamiento üìå

```python
# Eliminar filas con valores nulos en 'Name' y 'Genre'
df.dropna(subset=['Name', 'Genre'], inplace=True)
```

```python
# Calcular la mediana de la columna 'Year_of_Release'
mediana_year_of_release = df['Year_of_Release'].median()

# Imputar los valores nulos con la mediana
df['Year_of_Release'].fillna(mediana_year_of_release, inplace=True)
```

```python
# Calcular la moda de la columna 'Publisher'
moda_publisher = df['Publisher'].mode()[0]

# Imputar los valores nulos con la moda
df['Publisher'].fillna(moda_publisher, inplace=True)
```

```python
# Lista de columnas con valores nulos que deseas eliminar
columnas_con_nulos = ['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Developer', 'Rating']

# Eliminar las columnas con valores nulos
df = df.drop(columnas_con_nulos, axis=1)
```

### Detecci√≥n de Colinealidad üìå‚Äã

```python
test_df = df.copy()

test_df['new_global_sales'] = round(test_df['NA_Sales'] + test_df['EU_Sales'] + test_df['JP_Sales'] + test_df['Other_Sales'], 0)
test_df['Global_Sales'] = round(test_df['Global_Sales'], 0)

# Compara las ventas globales originales con la suma de las ventas por regi√≥n
comparison = test_df.loc[test_df['new_global_sales'] == test_df['Global_Sales'], ['Global_Sales', 'new_global_sales']]
print(comparison)
```

* Las ventas globales son iguales a la suma de las ventas por regi√≥n para una gran cantidad de registros (6924). Esto indica que las ventas por regi√≥n y las ventas globales est√°n relacionadas de manera directa, lo que justifica la colinealidad. Dado que las ventas por regi√≥n parecen ser una variable redundante en este contexto, podr√≠amos considerar eliminarlas de los datos.

```python
df = df.drop(['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], axis=1)
```

```python
# Visualizaci√≥n de un boxplot para 'Global_Sales'
plt.figure(figsize=(15, 3))
sns.boxplot(x=df['Global_Sales'], color='lightblue')
plt.title('Distribuci√≥n de Global_Sales')
plt.xlabel('Ventas Globales')
plt.show()

# Estad√≠sticos descriptivos de 'Global_Sales'
stats = df['Global_Sales'].describe()

# Valores at√≠picos (outliers)
q1 = stats['25%']
q3 = stats['75%']
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df['Global_Sales'] < lower_bound) | (df['Global_Sales'] > upper_bound)]

# Muestra de los valores at√≠picos
print("Valores at√≠picos (outliers) en 'Global_Sales':")
print(outliers)
```

<div style="border-radius:10px; border:#808080 solid; padding: 15px; background-color: ##F0E68C ; font-size:100%; text-align:left">

<h3 align="left"><font color=orange>üìé‚Äã Conclusi√≥n:</font></h3>

* La mayor√≠a de los juegos se encuentran en el rango de 0.2 a 20, luego hay una cantidad significativamente menor de juegos en el rango de 20 a 40. No hay puntos entre 41 y 80, excepto un valor √∫nico despu√©s de 80.

Esto podr√≠a permitirnos crear una divisi√≥n en tres categor√≠as basadas en cuartiles:

* Bajas Ventas: Ventas globales menores o iguales a 20.

* Ventas Moderadas: Ventas globales entre 20 y 40.

* Altas Ventas: Ventas globales mayores de 40.

```python
# Calcula los cuartiles
primer_cuartil = np.percentile(df['Global_Sales'], 25)
tercer_cuartil = np.percentile(df['Global_Sales'], 75)

# Define las categor√≠as
umbral_bajas_ventas = primer_cuartil
umbral_ventas_moderadas = tercer_cuartil
umbral_altas_ventas = np.max(df['Global_Sales'])

# Crea la columna 'Categoria_Ventas'
df['Categoria_Ventas'] = pd.cut(df['Global_Sales'],
                                bins=[-float('inf'), umbral_bajas_ventas, umbral_ventas_moderadas, umbral_altas_ventas, float('inf')],
                                labels=['Ventas Bajas', 'Ventas Moderadas', 'Ventas Altas', 'Ventas Muy Altas'])
```

```python
df
```

```python
# Contar la cantidad de juegos en cada categor√≠a de ventas
ventas_por_categoria = df['Categoria_Ventas'].value_counts()

# Ordenar las categor√≠as de ventas
ventas_por_categoria = ventas_por_categoria.reindex(['Ventas Bajas', 'Ventas Moderadas', 'Ventas Altas', 'Ventas Muy Altas'])

# Establecer el estilo y la paleta personalizada
sns.set(style="whitegrid")

# Crear un gr√°fico de barras utilizando Seaborn
plt.figure(figsize=(8, 3))
sns.barplot(x=ventas_por_categoria.index, y=ventas_por_categoria.values)
plt.title('Distribuci√≥n de Juegos por Categor√≠a de Ventas')
plt.xlabel('Categor√≠as')
plt.ylabel('Cantidad de Juegos')
plt.xticks(rotation=45)

# Mostrar el gr√°fico
plt.show()
```

Podr√≠amos preguntarnos si existe una relaci√≥n entre la frecuencia y las ventas globales, es decir, **¬øson los g√©neros con m√°s juegos los que m√°s se venden?**

## VALIDACION CRUZADA

A continuaci√≥n realizamos validaci√≥n cruzada con **5 folds** en el conjunto de datos calculando la precisi√≥n del modelo de Regresi√≥n Log√≠stica en cada fold, as√≠ como la precisi√≥n promedio.

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

# Inicializar el modelo de Regresi√≥n Log√≠stica
logistic_regression = LogisticRegression(random_state=42)

# Realizar validaci√≥n cruzada con 5 folds
scores = cross_val_score(logistic_regression, X, y, cv=5, scoring='accuracy')

print("Precisi√≥n por fold:", scores)
print("Precisi√≥n promedio:", scores.mean())
```

Los resultados muestran una precisi√≥n constante en cada uno de los folds y una precisi√≥n promedio bastante similar al modelo de Regresi√≥n Log√≠stica (74.85%).

Para evaluar el n√∫mero √≥ptimo de clusters usamos el m√©todo del codo, que en el gr√°fico representa el punto donde la m√©trica deja de disminuir r√°pidamente.

```python
# Lista para almacenar los valores de inercia
inertia = []

# Rango de n√∫mero de clusters a probar
range_clusters = range(2, 11)

# Calcular la inercia para cada n√∫mero de clusters
for num_clusters in range_clusters:
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(df_dummies)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(range_clusters, inertia, marker='o')
plt.xlabel('N√∫mero de Clusters')
plt.ylabel('Inercia')
plt.title('M√©todo del Codo (Elbow Method)')
plt.show()
```

El n√∫mero √≥ptimo de clusters deber√≠a estar entre 4 y 6.

```python
# Lista para almacenar los valores de la m√©trica de la silueta
silhouette_scores = []

# Rango de n√∫mero de clusters a probar
range_clusters = range(4, 7)  # Desde 4 hasta 6 clusters

# Calcular el score de Silhouette para cada n√∫mero de clusters
for num_clusters in range_clusters:
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(df_dummies)
    labels = kmeans.labels_
    silhouette_scores.append(silhouette_score(df_dummies, labels))

plt.figure(figsize=(8, 6))
plt.plot(range_clusters, silhouette_scores, marker='o')
plt.xlabel('N√∫mero de Clusters')
plt.ylabel('Puntaje de Silhouette')
plt.title('M√©todo de la Silueta (Silhouette Method)')
plt.show()
```

```python
# Calcula el coeficiente de Silhouette para cada muestra
silhouette_vals = silhouette_samples(df_dummies, kmeans.labels_)

# Graficar el gr√°fico de Silhouette
plt.figure(figsize=(14, 6))

y_lower, y_upper = 0, 0
yticks = []

for i in range(6):
    # Recopilar los coeficientes de Silhouette para las muestras pertenecientes a un cluster
    cluster_silhouette_vals = silhouette_vals[kmeans.labels_ == i]
    cluster_silhouette_vals.sort()

    # Calcular los l√≠mites inferior y superior para el cluster i
    y_upper += len(cluster_silhouette_vals)
    plt.barh(range(y_lower, y_upper), cluster_silhouette_vals, edgecolor='none', height=1)
    yticks.append((y_lower + y_upper) / 2)
    y_lower += len(cluster_silhouette_vals)

# L√≠nea que muestra el promedio global de los coeficientes de Silhouette
plt.axvline(np.mean(silhouette_vals), color='red', linestyle='--', linewidth=1, label='Promedio Global')
plt.yticks(yticks, range(6))
plt.xlabel('Coeficiente de Silhouette')
plt.ylabel('Cluster')
plt.title('Gr√°fico de Silhouette para 6 Clusters')
plt.legend()
plt.show()
```
